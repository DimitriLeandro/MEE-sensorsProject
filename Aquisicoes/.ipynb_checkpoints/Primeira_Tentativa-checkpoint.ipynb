```python
for i in range(0, 9):
    dividendo = absolute(app9[i] - encoder9[i])
    divisor = sqrt((2*ictz9[i])**2 + (2*ictzEnc9[i])**2)

    erroNormalizado = dividendo / divisor

    print("Erro Normalizado para " + str(encoder9[i]) + "º:\t\t" + str(erroNormalizado) + "\n")
```
